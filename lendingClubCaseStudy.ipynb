## Observation: Impact of Interest Rate on Loan Default

### The analysis reveals a clear relationship between the interest rate and the default rate. As the interest rate increases, the default rate percentage also rises significantly. Borrowers with higher interest rates are more likely to default on their loans compared to those with lower rates.

### 
Standardize this column to ensure consistency. Convert values like "10+ years", "<1 year", etc., into numerical values.

Handle missing values if they exist.

Example:

"10+ years" → 10

"5 years" → 5

"<1 year" → 0

"n/a" → NaN or a default value.

###

```python
# Clean 'emp_length' column
def clean_emp_length(value):
    if pd.isnull(value) or value.lower() == 'n/a':
        return None
    if '< 1' in value:
        return 0
    if '10+' in value:
        return 10
    return int(''.join(filter(str.isdigit, value)))

cleaned_loan_data['emp_length_cleaned'] = cleaned_loan_data['emp_length'].apply(clean_emp_length)

# Clean 'emp_title' column
cleaned_loan_data['emp_title_cleaned'] = cleaned_loan_data['emp_title'].str.strip().str.lower()

# Analyze top 10 job titles and default rates
top_titles = (
    cleaned_loan_data['emp_title_cleaned']
    .value_counts()
    .head(10)
    .index
)
title_default_rates = cleaned_loan_data[cleaned_loan_data['emp_title_cleaned'].isin(top_titles)].groupby(
    'emp_title_cleaned')['loan_status'].apply(lambda x: (x == 'Charged Off').mean() * 100)

# Analyze default rate by employment length
emp_length_default_rates = cleaned_loan_data.groupby('emp_length_cleaned')['loan_status'].apply(
    lambda x: (x == 'Charged Off').mean() * 100
).sort_index()
```

### Observation: Impact of DTI on Loan Default

##### The analysis shows that as the Debt-to-Income (DTI) ratio increases, the default rates also tend to increase. Borrowers with higher DTI ratios might face financial challenges in managing their existing debts and loans, making them more likely to default.

#### Low DTI (e.g., 0-10): Lower default rates due to manageable debt levels.
#### High DTI (e.g., 20+): Higher default rates, indicating financial strain.

```python
# Create bins for DTI
dti_bins = pd.cut(
    cleaned_loan_data['dti'], 
    bins=[0, 5, 10, 15, 20, 25, 30], 
    labels=['0-5', '5-10','10-15', '15-20','20-25', '25+']
)

# Add the bins to the dataset
cleaned_loan_data['dti_bin'] = dti_bins

# Calculate default rate for each DTI bin
dti_default_rates = cleaned_loan_data.groupby('dti_bin')['loan_status'].apply(
    lambda x: (x == 'Charged Off').mean() * 100
).sort_index()

# Plot default rates by DTI bin
plt.figure(figsize=(10, 6))
dti_default_rates.plot(kind='bar', color='orange', logy=True)
plt.title('Default Rates by DTI Range')
plt.xlabel('DTI Range')
plt.ylabel('Default Rate (%)')
plt.tight_layout()
plt.show()
```

### Observation Default Percentage against issued Loans.

### Clearly we can see 2007 has highest Default Rate. The 2007 financial crisis might have led to higher defaults

```python
 # Step 1: Convert 'issue_d' to datetime format
cleaned_loan_data['issue_d'] = pd.to_datetime(cleaned_loan_data['issue_d'], format='%b-%y', errors='coerce')

# Step 2: Extract year from issue date
cleaned_loan_data['issue_year'] = cleaned_loan_data['issue_d'].dt.year

# Step 3: Group by year to calculate loans issued and defaults
yearly_data = cleaned_loan_data.groupby('issue_year').agg(
    loans_issued=('loan_status', 'count'),
    loans_defaulted=('loan_status', lambda x: (x == 'Charged Off').sum())
)

# Step 4: Calculate default percentages
yearly_data['default_percentage'] = (yearly_data['loans_defaulted'] / yearly_data['loans_issued']) * 100

# Display the results
print(yearly_data)

# Step 5: Plot default percentage
plt.figure(figsize=(10, 6))
yearly_data['default_percentage'].plot(kind='line', marker='o', color='red', label='Default Percentage')
plt.title('Default Percentage by Year')
plt.xlabel('Year')
plt.ylabel('Default Percentage (%)')   
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```
